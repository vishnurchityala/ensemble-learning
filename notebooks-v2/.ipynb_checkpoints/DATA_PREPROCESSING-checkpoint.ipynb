```python
import os
import pandas as pd
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
benign_files = os.listdir("../datasetv2/benign/")
malignant_files = os.listdir("../datasetv2/malignant/")

print("Number of Benign samples:",len(benign_files))
print("Number of Malignant samples:",len(malignant_files))
```

```python
random.shuffle(benign_files)
random.shuffle(malignant_files)
```

```python
benign_split = int(len(benign_files) * 0.8)
malignant_split = int(len(malignant_files) * 0.8)
```

```python
benign_train = benign_files[:benign_split]
benign_test = benign_files[benign_split:]
malignant_train = malignant_files[:malignant_split]
malignant_test = malignant_files[malignant_split:]
```

```python
metadata = []
```

```python
for path in benign_train:
    metadata.append({"image_path": path, "label": "benign", "split": "train"})
for path in benign_test:
    metadata.append({"image_path": path, "label": "benign", "split": "test"})

for path in malignant_train:
    metadata.append({"image_path": path, "label": "malignant", "split": "train"})
for path in malignant_test:
    metadata.append({"image_path": path, "label": "malignant", "split": "test"})
```

```python
df = pd.DataFrame(metadata)
```

```python
df["image_path"] = df["label"] + "/" + df["image_path"]
df
```

```python
df.to_csv("./CSV/dataset_metadata.csv", index=False)

print("Dataset Metadata CSV file created successfully!")
```

```python
def process_image(input_path, output_path, size=(224, 224)):
    """
    Reads an image, normalizes pixel values, resizes it, 
    applies CLAHE (Contrast Limited Adaptive Histogram Equalization), 
    and saves the processed image.

    Args:
        input_path (str): Path to the input image.
        output_path (str): Path to save the processed image.
        size (tuple): Desired output image size (default: 256x256).
    """
    image = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print(f"Error: Unable to load image {input_path}")
        return

    image = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)

    image = cv2.resize(image, size, interpolation=cv2.INTER_AREA)

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    image = clahe.apply(image)

    cv2.imwrite(output_path, image)

    print(f"Processed image saved to {output_path}")
```

```python
input_image = "../datasetv2/malignant/Mass-Training_P_00444_LEFT_MLO.png"
output_image = "Mass-Training_P_00444_LEFT_MLO.png"

original_img = cv2.imread(input_image, cv2.IMREAD_GRAYSCALE)
process_image(input_image,output_image)
processed_img = cv2.imread(output_image, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_img, cmap="gray")
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(processed_img, cmap="gray")
plt.title("Processed Image (CLAHE + Normalized)")
plt.axis("off")

plt.show()
```
