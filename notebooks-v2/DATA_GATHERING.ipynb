```python
# Data Pre-Processing
import os
import pandas as pd

benign_directory = "../../DATASET/dataset/BENIGN/manifest-1742213063255"
malignant_directory = "../../DATASET/dataset/MALIGNANT/manifest-1742233833522"
```

```python
metadata_benign = pd.read_csv("metadata-benign.csv")
metadata_benign
```

```python
metadata_benign["File Location"] = metadata_benign["File Location"].str.lstrip(".")
metadata_benign
```

```python
for file in metadata_benign["File Location"]:
    print(benign_directory+file)
```

```python
import pydicom
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def dcm_to_png(dcm_path, png_path):
    dicom_data = pydicom.dcmread(dcm_path)
    pixel_array = dicom_data.pixel_array
    pixel_array = pixel_array - np.min(pixel_array)
    pixel_array = (pixel_array / np.max(pixel_array) * 255).astype(np.uint8)
    image = Image.fromarray(pixel_array)
    image.save(png_path)
#     plt.imshow(image, cmap="gray")
#     plt.axis("off")
#     plt.title("Converted PNG Image")
#     plt.show()
    print(f"Converted {dcm_path} to {png_path}")

output_directory = "../datasetv2/"
not_found_count = 0  # Counter for missing files

for file in metadata_benign["File Location"]:
    dcm_path = benign_directory + file + "/1-1.dcm"
    output_png = "../datasetv2/" + file.split('/')[2]+".png"
    print(f"Converting {dcm_path} to PNG {output_png}...")
    dcm_to_png(dcm_path, output_png)
    print(f"Converted {dcm_path} to PNG {output_png}")
```
