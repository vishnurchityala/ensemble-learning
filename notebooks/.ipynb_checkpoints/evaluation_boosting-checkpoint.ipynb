```python
import tensorflow as tf
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
import xgboost as xgb
import lightgbm as lgb
```

```python
print(f"TensorFlow: {tf.__version__}")
print(f"NumPy: {np.__version__}")
print(f"OpenCV: {cv2.__version__}")
print(f"Pandas: {pd.__version__}")
```

```python
metadata = pd.read_csv("../augmented_dataset/augmented_metadata.csv")
```

```python
metadata
```

```python
split_label_counts = metadata.groupby(['split', 'label']).size().reset_index(name='count')

print(split_label_counts)
```

```python
train_data = metadata[metadata['split'] == 'train']
train_data
```

```python
test_data = metadata[metadata['split'] == 'test']
test_data
```

```python
def load_image(image_path, target_size=(224, 224)):
    image = tf.io.read_file(image_path) 
    image = tf.image.decode_png(image, channels=3)
    image = tf.image.resize(image, target_size)
    image = image / 255.0 
    return image.numpy()

def load_mask(mask_path, target_size=(224, 224)):
    mask = tf.io.read_file(mask_path)
    mask = tf.image.decode_png(mask, channels=1)
    mask = tf.image.resize(mask, target_size) 
    mask = mask / 255.0
    return mask.numpy()

X_train = np.array([load_image("../augmented_dataset/"+image_path) for image_path in train_data['image_path']])
mask_train = np.array([load_mask("../augmented_dataset/"+mask_path) for mask_path in train_data['mask_path']])
y_train = train_data['label'].values

X_test = np.array([load_image("../augmented_dataset/"+image_path) for image_path in test_data['image_path']])
mask_test = np.array([load_mask("../augmented_dataset/"+mask_path) for mask_path in test_data['mask_path']])
y_test = test_data['label'].values

def display_samples(images, masks, labels, dataset_type="Train", num_samples=2):
    fig, axes = plt.subplots(num_samples, 2, figsize=(10, num_samples * 3))
    fig.suptitle(f"{dataset_type} Set Samples", fontsize=16, fontweight="bold")

    for i in range(num_samples):
        axes[i, 0].imshow(images[i])
        axes[i, 0].set_title(f"Image - Label: {labels[i]}")
        axes[i, 0].axis("off")

        axes[i, 1].imshow(masks[i].squeeze(), cmap="gray")
        axes[i, 1].set_title("Mask")
        axes[i, 1].axis("off")

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

display_samples(X_train, mask_train, y_train, dataset_type="Train")
display_samples(X_test, mask_test, y_test, dataset_type="Test")
```

```python
def load_fe(feature_extractor):
    return tf.keras.applications.__dict__[feature_extractor](
        include_top=False, weights='imagenet', input_shape=(224, 224, 3), pooling='avg'
    )
feature_extractors = {
    "DenseNet201": load_fe("DenseNet201"),
    "InceptionV3": load_fe("InceptionV3"),
    "MobileNetV2": load_fe("MobileNetV2"),
}

print(feature_extractors)
```

```python
for fe_name, fe_model in feature_extractors.items():
    print(f"Basic Summary for {fe_name}:")
    num_params = fe_model.count_params()
    num_layers = len(fe_model.layers)
    print(f"Total Parameters: {num_params}")
    print(f"Number of Layers: {num_layers}")
    print("\n")
```

```python
boosting_methods = {
    "ADB": AdaBoostClassifier,
    "GB": GradientBoostingClassifier,
    "XGB": xgb.XGBClassifier,
    "LGBM": lgb.LGBMClassifier
}
```

```python
tree_counts = [50, 100, 150, 200]
```

```python
model_dir = "models"
os.makedirs(model_dir, exist_ok=True)
```

```python
extracted_features= {}
```

```python
for fe_name, fe_model in feature_extractors.items():
    print(f"Extracting Features Using : {fe_name}")
    X_features = fe_model.predict(X_train)
    extracted_features[fe_name] = X_features
```

```python
extracted_features_test = {}

for fe_name, fe_model in feature_extractors.items():
    print(f"Extracting Test Features Using : {fe_name}")
    X_test_features = fe_model.predict(X_test)
    extracted_features_test[fe_name] = X_test_features
```

```python
model_dir = "models"
os.makedirs(model_dir, exist_ok=True)

model_paths = []

for fe_name, X_train_features in extracted_features.items():
    X_test_features = extracted_features_test[fe_name]

    print(f"Training Decision Tree Classifier for {fe_name}")

    model = DecisionTreeClassifier()
    model.fit(X_train_features, y_train)

    model_filename = f"{model_dir}/dt_{fe_name}.pkl"
    with open(model_filename, 'wb') as f:
        pickle.dump(model, f)
    
    model_paths.append({"FeatureExtractor": fe_name, "ModelPath": model_filename})

model_paths_df = pd.DataFrame(model_paths)
model_paths_df.to_csv("dt_model_paths.csv", index=False)

print("\n All models saved as .pkl and paths stored in 'dt_model_paths.csv'")
```

```python
from tabulate import tabulate 

results = []

for model_info in model_paths:
    fe_name = model_info["FeatureExtractor"]
    model_filename = model_info["ModelPath"]

    with open(model_filename, 'rb') as f:
        model = pickle.load(f)

    X_test_features = extracted_features_test[fe_name]
    y_pred = model.predict(X_test_features)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    results.append([fe_name, accuracy, precision, recall, f1])

print("\n Model Evaluation Results:\n")
print(tabulate(results, headers=["Feature Extractor", "Accuracy", "Precision", "Recall", "F1 Score"], tablefmt="grid"))
```

```python
import pickle
import os
import pandas as pd
from sklearn.preprocessing import LabelEncoder

model_dir = "models"
os.makedirs(model_dir, exist_ok=True)

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

with open(f"{model_dir}/label_encoder.pkl", "wb") as f:
    pickle.dump(label_encoder, f)

n_estimators_list = [50, 100, 150, 200]
model_paths_bt = []

for fe_name, X_train_features in extracted_features.items():
    X_test_features = extracted_features_test[fe_name]

    for boost_name, BoostingModel in boosting_methods.items():
        for n_trees in n_estimators_list:
            print(f"\n Training {boost_name} with {n_trees} trees for {fe_name}")

            model = BoostingModel(n_estimators=n_trees)
            model.fit(X_train_features, y_train_encoded)

            model_filename = f"{model_dir}/{fe_name}_{boost_name}_{n_trees}.pkl"
            with open(model_filename, "wb") as f:
                pickle.dump(model, f)

            model_paths_bt.append({
                "FeatureExtractor": fe_name,
                "Model": boost_name,
                "Trees": n_trees,
                "ModelPath": model_filename
            })

model_paths_df = pd.DataFrame(model_paths_bt)
model_paths_df.to_csv(f"boosting_model_paths.csv", index=False)

print("\n All models saved and paths stored in 'boosting_model_paths.csv'")
```

```python
model_dir = "models"
model_paths_df = pd.read_csv(f"boosting_model_paths.csv")

y_test = label_encoder.fit_transform(y_test)


evaluation_results = []

for _, row in model_paths_df.iterrows():
    fe_name = row["FeatureExtractor"]
    boost_name = row["Model"]
    n_trees = row["Trees"]
    model_path = row["ModelPath"]

    print(f"\n Evaluating {boost_name} with {n_trees} trees for {fe_name}")

    with open(model_path, 'rb') as f:
        model = pickle.load(f)

    X_test_features = extracted_features_test[fe_name]

    y_pred = model.predict(X_test_features)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    evaluation_results.append({
        "FeatureExtractor": fe_name,
        "Model": boost_name,
        "Trees": n_trees,
        "Accuracy": round(accuracy, 4),
        "Precision": round(precision, 4),
        "Recall": round(recall, 4),
        "F1 Score": round(f1, 4)
    })

evaluation_df = pd.DataFrame(evaluation_results)

evaluation_df.to_csv(f"boosting_model_evaluation.csv", index=False)

print("\n Evaluation complete! Results saved in 'boosting_model_evaluation.csv'")
```
