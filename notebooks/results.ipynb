```python
import pickle
from tabulate import tabulate

with open("bootstrapped_hybrid_architectures.pkl", "rb") as file:
    bootstrapped_hybrid_architectures = pickle.load(file)

table_data = []
for model, bootstraps in bootstrapped_hybrid_architectures.items():
    for bootstrap, files in bootstraps.items():
        table_data.append([model, bootstrap, "\n".join(files)])

headers = ["Model", "Bootstrap", "Files"]

print("\n Bootstrapped Hybrid Ensemble Models\n")

print(tabulate(table_data, headers=headers, tablefmt="grid"))
```
